```python
import pandas as pd
df = pd.read_csv('../data/mahomes.csv')
df
```

```python
df.plot.scatter(x='YDS', y = 'TD')
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(df.iloc[:,5:6], df.TD)
```

```python
reg.score(df.iloc[:,5:6], df.TD)
```

```python
import matplotlib.pyplot as plt

plt.scatter(df.YDS, df.TD, color='blue')
plt.plot(df.YDS, reg.predict(df.iloc[:,5:6]), color='black')

plt.show()
```
