# Result Plotting

Here we read in the csv that contains our prediction results, and we can modify the last cell to plot whatever we want for visual comparison!

```python
import pandas as pd

df = pd.read_csv("QB_2022_results.csv")

df=df.iloc[:-1,:]

df
```

```python
#1 actual - blue
#2 espn - green
#3 lr 2x - red
#4 lr - standard - red
#5 lasso standard - orange
#6 ridge standard - yellow
#7 lasso 2x - orange
#8 lasso all data - orange
#9 lasso def weights - orange
#10 lasso home/away - orange


import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(18,11)})

actual = sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,1], label="Actual")
espn = sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,2], label="ESPN", color='Green')
espn = sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,4], label="LR", color='Red')
espn = sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,5], label="Lasso", color='Orange')
espn = sns.lineplot(x=df.iloc[:,0], y=df.iloc[:,6], label="Ridge", color='Yellow')
plt.ylabel("Points", size=20)
plt.title("Actual Points Vs. All Prediction Models", size=40)
actual.legend(fontsize=20)
plt.xticks(rotation=90)
plt.show()
```
