```python
import pandas as pd

df = pd.read_csv("../data/passing.tsv", sep='\t')
```

```python
df
```

```python
import matplotlib.pyplot as plt
df = df.head(32)
plt.figure(figsize=(40,20))
plt.scatter(df['Player'], df['Yds'])
plt.xlabel('Player')
plt.ylabel('Passing Yards')
plt.title('Passing Yards per Player')
plt.show()
```

```python
df = df.head(16)
plt.figure(figsize=(30,20))
plt.scatter(df['Player'], df['YPA'])
plt.xlabel('Player')
plt.ylabel('Passing Yards')
plt.title('Passing Yards per Player')
plt.show()
```
